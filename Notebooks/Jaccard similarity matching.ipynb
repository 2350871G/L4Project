```python
import os
import srsly

TYN_DUMP_PATH = r'/content/drive/MyDrive/4th year project/tyn_dump - Copy'

articles = os.listdir(TYN_DUMP_PATH)

a = srsly.read_json(os.path.join(TYN_DUMP_PATH, articles[19]))
print(a['url'])
text = a['text']
tyn = a['tyn']

import spacy
import en_core_web_sm
nlp = en_core_web_sm.load()

doc = nlp(text)

noun_chunks = doc.noun_chunks

tyn_docs = [nlp(t) for t in tyn]

def jaccard(doc1, doc2):
  s1 = set([t.lower for t in doc1])
  s2 = set([t.lower for t in doc2])
  return (len(s1.intersection(s2)))/(len(s1.union(s2)))

for noun in noun_chunks:
  for thing in tyn_docs:
    j = jaccard(noun, thing)
    if j > 0.5:
      print(j, "||", noun, "||", thing)
```
